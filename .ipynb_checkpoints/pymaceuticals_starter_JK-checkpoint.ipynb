```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem


# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data_df = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_data_df = pd.read_csv("data/clinicaltrial_data.csv")
#print(mouse_data_df.head())
#print(clinical_trial_data_df.head())
#print(mouse_data_df.tail())
#print(clinical_trial_data_df.tail())

# Combine the data into a single dataset
combined_data_df = pd.merge(mouse_data_df, clinical_trial_data_df, on="Mouse ID", how="outer")

# Display the data table for preview
combined_data_df.head()
#combined_data_df.count()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Store as a dataframe
grouped_combined_data_df = combined_data_df.groupby(['Drug', 'Timepoint'])
mean_grouped_combined_data_df = grouped_combined_data_df.mean()

# Preview DataFrame
mean_grouped_combined_data_df.head()
```

```python
#Reset the index
mean_grouped_combined_data_df_reset_index = mean_grouped_combined_data_df.reset_index()
mean_grouped_combined_data_df_reset_index
#Drop the column "Metastatic Sites"
reorganized_tumor_vol_df = mean_grouped_combined_data_df_reset_index.drop(columns=['Metastatic Sites'])
reorganized_tumor_vol_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Create Data Frame
# Preview DataFrame
tumor_volume_error = grouped_combined_data_df.sem()
tumor_volume_error_df = tumor_volume_error
tumor_volume_error_df
# Drop columns 'Metastatic Sites' and 'Mouse ID'                                                                            
reorganized_tumor_vol_df_drop = tumor_volume_error_df.drop(columns=['Metastatic Sites', 'Mouse ID'])
reorganized_tumor_vol_df_drop
# Reset Index
reorganized_tumor_vol_df_drop_reset = reorganized_tumor_vol_df_drop.reset_index()
reorganized_tumor_vol_df_drop_reset.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames Transpose
# Preview that Reformatting worked
transpose_pivot_table = reorganized_tumor_vol_df.pivot_table(index=['Timepoint'], columns=['Drug'],
                     values='Tumor Volume (mm3)')
transpose_pivot_table
```

```python
# Minor Data Munging to Re-Format for SEM data frame Transpose
# Preview that Reformatting worked
transpose_pivot_table_sem = reorganized_tumor_vol_df_drop_reset.pivot_table(index=['Timepoint'], columns=['Drug'],
                     values='Tumor Volume (mm3)', aggfunc='first', fill_value=0)
transpose_pivot_table_sem
```

```python
print(transpose_pivot_table.index.values)
print(transpose_pivot_table.columns)
```

```python
# Generate the Plot (with Error Bars)
# Plot error bars in same color and overlay line graph
ax = transpose_pivot_table[["Capomulin", "Infubinol", "Ketapril", "Placebo"]].plot(figsize=(8,6), ls = 'none', yerr = transpose_pivot_table_sem[["Capomulin", "Infubinol", "Ketapril", "Placebo"]], color = ['m', 'b', 'g', 'k'], legend = False, linewidth=0.5)
ax.set_prop_cycle(None)
transpose_pivot_table[["Capomulin", "Infubinol", "Ketapril", "Placebo"]].plot(figsize=(8,6), style=['mo:', 'b^:', 'gs:', 'kd:'], ax=ax)

#Make it look nice
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.xlim(-2, 47)
plt.ylim(33, 74)
plt.grid()
#plt.legend(title='')

# Save the Figure
plt.savefig("images/tumor_response_to_treatment.png")
```

```python
# Show the Figure
plt.show()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrame
metastatic_error = grouped_combined_data_df.sem()
metastatic_error_df = metastatic_error
metastatic_error_df
                                                                            
metastatic_error_drop_df = metastatic_error_df.drop(columns=['Tumor Volume (mm3)', 'Mouse ID'])
metastatic_error_drop_df

metastatic_error_drop_df_reset = metastatic_error_drop_df.reset_index()
metastatic_error_drop_df_reset.head()
```

```python
# Generate the Plot (with Error Bars)
# Save the Figure
# Show the Figure
ax = transpose_pivot_table_metastatic[["Capomulin", "Infubinol", "Ketapril", "Placebo"]].plot(figsize=(8,6), ls = 'none', yerr = transpose_pivot_table_metastatic_sem[["Capomulin", "Infubinol", "Ketapril", "Placebo"]], color = ['m', 'b', 'g', 'k'], legend = False, linewidth=0.5)
ax.set_prop_cycle(None)
transpose_pivot_table_metastatic[["Capomulin", "Infubinol", "Ketapril", "Placebo"]].plot(figsize=(8,6), style=['mo:', 'b^:', 'gs:', 'kd:'], ax=ax)

#Make it look nice
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.xlim(-2, 47)
plt.ylim(-0.2, 3.8)
plt.grid()
#plt.legend(title='')

# Save the Figure
plt.savefig("images/Metastatic Spread During Treatment.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame
grouped_combined_data_df_drug_time = combined_data_df.groupby(['Drug', 'Timepoint'])
count_grouped_combined_data_df = grouped_combined_data_df_drug_time.count()

# Preview DataFrame
count_grouped_combined_data_df.head()
```

```python
count_grouped_combined_data_df_drop = count_grouped_combined_data_df.drop(columns=['Tumor Volume (mm3)', 'Metastatic Sites'])
count_grouped_combined_data_df_drop

count_grouped_combined_data_df_drop_rename = count_grouped_combined_data_df_drop.rename(columns={"Mouse ID": "Mouse Count"})
count_grouped_combined_data_df_drop_rename

count_grouped_combined_data_df_drop_rename_reset = count_grouped_combined_data_df_drop_rename.reset_index()
count_grouped_combined_data_df_drop_rename_reset.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# Preview the Data Frame
count_grouped_combined_data_df_drop_rename_reset_transpose = count_grouped_combined_data_df_drop_rename_reset.pivot_table(index=['Timepoint'], columns=['Drug'],
                     values='Mouse Count', aggfunc='first', fill_value=0)
count_grouped_combined_data_df_drop_rename_reset_transpose
```

```python
# Generate the Plot (Accounting for percentages)
#column_headers = list(count_grouped_combined_data_df_drop_rename_reset_transpose)

#survival_rate = []

#i = 0
#j = 0
#for i in range(len(count_grouped_combined_data_df_drop_rename_reset_transpose)):
    #percentage_survival = ((count_grouped_combined_data_df_drop_rename_reset_transpose.iloc[i, 0]) / (count_grouped_combined_data_df_drop_rename_reset_transpose.iloc[0, j]))
   #survival_rate.append(percentage_survival)
#j = j + 1
#for j in range(len(count_grouped_combined_data_df_drop_rename_reset_transpose)):
    #j = j + 1
            #survival_rate.append(percentage_survival)
#print(survival_rate)

#df = pd.DataFrame(np.array(survival_rate).reshape(10,10), columns=column_headers)
#print (df)

# Save the Figure
# Show the Figure
#plt.show()
```

```python
# Generate the Plot (Accounting for percentages)
percent_survival_capomulin = ((count_grouped_combined_data_df_drop_rename_reset_transpose['Capomulin']) / (max(count_grouped_combined_data_df_drop_rename_reset_transpose['Capomulin']))) *100
percent_survival_capomulin

percent_survival_Infubinol = ((count_grouped_combined_data_df_drop_rename_reset_transpose['Infubinol']) / (max(count_grouped_combined_data_df_drop_rename_reset_transpose['Infubinol']))) *100
percent_survival_Infubinol

percent_survival_Ketapril = ((count_grouped_combined_data_df_drop_rename_reset_transpose['Ketapril']) / (max(count_grouped_combined_data_df_drop_rename_reset_transpose['Ketapril']))) *100
percent_survival_Ketapril

percent_survival_Placebo = ((count_grouped_combined_data_df_drop_rename_reset_transpose['Placebo']) / (max(count_grouped_combined_data_df_drop_rename_reset_transpose['Placebo']))) *100
percent_survival_Placebo

df = pd.DataFrame()
df['Capomulin'] = percent_survival_capomulin
df['Infubinol'] = percent_survival_Infubinol
df['Ketapril'] = percent_survival_Ketapril
df['Placebo '] = percent_survival_Placebo 

percentage_survivial = pd.DataFrame(df)
percentage_survivial
```

```python
percentage_survivial.plot(figsize=(8,6), style=['mo:', 'b^:', 'gs:', 'kd:'])
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(loc="lower left")

plt.xlim(-2, 47)
plt.ylim(31,103)
plt.grid()

# Save the Figure
plt.savefig("images/Survival During Treatment.png")
# Show the Figure
#plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
# Display the data to confirm
#Find the percent tumor volume change for the drugs: 'Capomulin','Infubinol','Ketapril','Placebo'
percent_tumor_volume_change = ((transpose_pivot_table.loc[45, :] - transpose_pivot_table.loc[0, :]) / transpose_pivot_table.loc[0, :] * 100)[['Capomulin','Infubinol','Ketapril','Placebo']]
percent_tumor_volume_change
#Find the percent tumor volume change for all the drugs
percent_tumor_volume_change_all = ((transpose_pivot_table.loc[45, :] - transpose_pivot_table.loc[0, :]) / transpose_pivot_table.loc[0, :] * 100)
percent_tumor_volume_change_all

#percent_tumor_volume_change_all_tuple = [(percent_tumor_volume_change_all)]
#percent_tumor_volume_change_all_tuple
```
